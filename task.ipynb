```python
import pandas as pd
```

```python
data = pd.DataFrame(pd.read_excel("/content/Study Title REV with Sample.xlsx"))
```

```python
data.isnull().sum()
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
causes=data['List of Causes']
title=data['Public Title of Study']
causes=causes.dropna(axis=0)
title=title.dropna(axis=0).tolist()
title[1]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
sentences = []
metrices=[]

sentence=[]
causeName=[]
match=[]
n=7
for x in range(len(causes)):
  sentences.append(causes[x])

for i in range(len(title)):
  sentences.append(title[i])
  sentence.append(title[i])
  sentence_embeddings = model.encode(sentences)
  metrices=cosine_similarity([sentence_embeddings[-1]],sentence_embeddings[:255])
  causearr=np.argmax(metrices)
  arr=np.sort(metrices)
  percn=np.amax(metrices)
  causeName.append(sentences[causearr])
  match.append(percn*100)


```

```python
finalDataFrame=pd.DataFrame(sentence)
finalDataFrame['Cause']=pd.DataFrame(causeName)
finalDataFrame['Match in %']=pd.DataFrame(match)
```
